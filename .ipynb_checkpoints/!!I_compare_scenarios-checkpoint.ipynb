```python
import os
import numpy as np
import pandas as pd
import winsound
from tkinter import Tcl
import glob
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import scipy.stats as stat
```

```python
##give each run a name

A='test 1'
B='test 2'
dir_name="Compare_scenarios_"+A+'_vs_'+B
dir_name
```

```python
##making a dir for comparison

cur_dir= os.getcwd()
if os.path.exists(os.path.join(cur_dir,'output',dir_name)) and os.path.isdir(os.path.join(cur_dir,'output',dir_name)):
    print("!!WARNING!!")
    print("Folder already exists!")
else:
    os.mkdir(os.path.join(cur_dir,'output',dir_name))
    print("creating folder")
```

```python
##defining save finction


def save_fig(_id, tight_layout=True, resolution=300):
    path = os.path.join(os.path.join(cur_dir,'output',dir_name), _id + ".png")
    print("Saving figure", _id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format='png', dpi=resolution)
```

```python
params=pd.read_excel('input/Cascade_info.xlsx',sheet_name='plotting_info')
name=params['name']
name=sorted(name)
name
```

```python
## path to result of each run
cur_dir=os.getcwd()
par_dir=os.path.dirname(cur_dir)
A_path=os.path.join("C:\\Users\\chrs-164-ip\\Desktop\\Projects\\SDDP\\General structure\\For github\\output","solution_scenarios")
B_path=os.path.join("C:\\Users\\chrs-164-ip\\Desktop\\Projects\\SDDP\\General structure\\For github\\output","solution_scenarios")
print(A_path,"\n",B_path)
```

```python
A_dict,B_dict={},{}
for dam in name:
    A_dict[dam]=""
    B_dict[dam]=""
```

```python
file_end=('release.csv','storage.csv')

##for run A
for n in name:
    temp_list=[]
    for file in os.listdir(A_path):
        file_path=os.path.join(A_path,file)
        if file.startswith(n) and file.endswith(file_end):
            temp=pd.read_csv(file_path,header=None)
            temp_list.append(temp)
        A_dict[n]=temp_list

##for run B
for n in name:
    temp_list=[]
    for file in os.listdir(B_path):
        file_path=os.path.join(B_path,file)
        if file.startswith(n) and file.endswith(file_end):
            temp=pd.read_csv(file_path,header=None)
            temp_list.append(temp)
        B_dict[n]=temp_list
```

```python
A_dict[name[0]][0].shape[1]
```

```python
A_dict
```

```python
n_sce=A_dict[name[0]][0].shape[1]

A_s=np.zeros(n_sce)
A_r=np.zeros(n_sce)
B_s=np.zeros(n_sce)
B_r=np.zeros(n_sce)

for n in name:
    for j in range(len(A_dict[name[0]])):
        a=A_dict[n][j].tail(1)
        b=B_dict[n][j].tail(1)
        if j==0:
            A_r=np.vstack((A_r,a))
            B_r=np.vstack((B_r,b))
        elif j==1:
            A_s=np.vstack((A_s,a))
            B_s=np.vstack((B_s,b))
        else:
            print("something wrong: list lenght is unexcpected!!")
A_r=A_r[1:]
A_s=A_s[1:]

B_r=B_r[1:]
B_s=B_s[1:]           
```

```python
A_r.shape
```

```python
release_list=[A_r,B_r]
storage_list=[A_s,B_s]
```

```python
xtick=name.copy()
for i in range(len(name)):
    xtick.append("")
```

```python
xtick
```

```python
##plotting box plots for release

positions =np.arange(1,1+len(name)*.6,.6)
fig=plt.figure(figsize=(40,20),dpi=300)
ax=fig.add_subplot(111,label='1')
plt.boxplot(release_list[0].T ,positions=positions,patch_artist=True,
           widths=.25,boxprops=dict(facecolor='b', color='b',alpha=.5,linewidth=5),
           medianprops=dict(color='k',linewidth=5),
       )
plt.boxplot(release_list[1].T ,positions=positions+.3,patch_artist=True,
           widths=.25,boxprops=dict(facecolor='r', color='r',alpha=.5,linewidth=5),
           medianprops=dict(color='g',linewidth=5)
       )
legend_smooth=mpatches.Patch(color='b',label=f'{A}')
legend_normal=mpatches.Patch(color='r',alpha=.5,label=f'{B}')
ax.legend(handles=[legend_smooth, legend_normal], fontsize=40)
plt.grid(True,which='major',axis='y')
ax.set_xticks(np.hstack((positions,positions+.25))+.1)
ax.set_xticklabels(xtick,fontsize=30)
ax.tick_params(axis='both', which='major', labelsize=30)
plt.title(f"Release: {A} vs. {B} scenarios",fontsize=40)
plt.xlim(.75,1+len(name)*.6)
plt.ylabel('CFs',fontsize=40)
ax.yaxis.get_offset_text().set_size(30)
save_fig('release compare scenarios')
```

```python
##plotting box plots for Storage

positions =np.arange(1,1+len(name)*.6,.6)
fig=plt.figure(figsize=(40,20),dpi=300)
ax=fig.add_subplot(111,label='1')
plt.boxplot(storage_list[0].T ,positions=positions,patch_artist=True,
           widths=.25,boxprops=dict(facecolor='b', color='b',alpha=.5,linewidth=5),
           medianprops=dict(color='k',linewidth=5),
       )
plt.boxplot(storage_list[1].T ,positions=positions+.3,patch_artist=True,
           widths=.25,boxprops=dict(facecolor='r', color='r',alpha=.5,linewidth=5),
           medianprops=dict(color='g',linewidth=5)
       )
legend_smooth=mpatches.Patch(color='b',label=f'{A}')
legend_normal=mpatches.Patch(color='r',alpha=0.5,label=f'{B}')
ax.legend(handles=[legend_smooth, legend_normal], fontsize=40)
plt.grid(True,which='major',axis='y')
ax.set_xticks(np.hstack((positions,positions+.25))+.1)
ax.set_xticklabels(xtick,fontsize=30)
ax.tick_params(axis='both', which='major', labelsize=30)
plt.title(f"Storage: {A} vs. {B} scenarios",fontsize=40)
plt.xlim(.75,1+len(name)*.6)
plt.ylabel('MAF',fontsize=40)
ax.yaxis.get_offset_text().set_size(30)
save_fig('Storage compare scenarios')
```
