```python
import os
import numpy as np
import pandas as pd
import winsound
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import scipy.stats as stat
```

```python
##give each run a name

A='normal'
B='smooth'
dir_name="Compare_adaptibility_"+A+'_vs_'+B
dir_name
```

```python
params=pd.read_excel('input/Cascade_info.xlsx',sheet_name='plotting_info')
names=params['name']
names=sorted(names)
names
```

```python
##making a dir for comparison

cur_dir= os.getcwd()
if os.path.exists(os.path.join(cur_dir,'output',dir_name)) and os.path.isdir(os.path.join(cur_dir,'output',dir_name)):
    print("!!WARNING!!")
    print("Folder already exists!")
else:
    os.mkdir(os.path.join(cur_dir,'output',dir_name))
    print("creating folder")
```

```python
##defining save finction


def save_fig(_id, tight_layout=True, resolution=300):
    path = os.path.join(os.path.join(cur_dir,'output',dir_name), _id + ".png")
    print("Saving figure", _id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format='png', dpi=resolution)
```

```python
## path to result of each run

A_path="C:\\Users\\chrs-164-ip\\Desktop\\Projects\\SDDP\\General structure\\For github\\output\\temp_files"
B_path="C:\\Users\\chrs-164-ip\\Desktop\\Projects\\SDDP\\General structure\\For github\\output\\temp_files"
print(A_path,"\n",B_path)
```

```python
##loading np array for each run

A_list=[]
for name in ["release","storage"]:
    file_n=os.path.join(A_path,name+"_boxplot_arr.npy")
    file=np.load(file_n)
    A_list.append(file)

B_list=[]
for name in ["release","storage"]:
    file_n=os.path.join(B_path,name+"_boxplot_arr.npy")
    file=np.load(file_n)
    B_list.append(file)
```

```python
xtick=names.copy()
for i in range(len(names)):
    xtick.append("")
xtick
```

```python
##plotting box plots for release

positions =np.arange(1,1+len(names)*.6,.6)
fig=plt.figure(figsize=(40,20),dpi=300)
ax=fig.add_subplot(111,label='1')
plt.boxplot(A_list[0].T ,positions=positions,patch_artist=True,
           widths=.25,boxprops=dict(facecolor='b', color='b',alpha=.5,linewidth=5),
           medianprops=dict(color='k',linewidth=5),
       )
plt.boxplot(B_list[0].T ,positions=positions+.3,patch_artist=True,
           widths=.25,boxprops=dict(facecolor='r', color='r',alpha=.5,linewidth=5),
           medianprops=dict(color='g',linewidth=5)
       )
legend_smooth=mpatches.Patch(color='b',label=f'{A}')
legend_normal=mpatches.Patch(color='r',label=f'{B}')
ax.legend(handles=[legend_smooth, legend_normal], fontsize=40)
plt.grid(True,which='major',axis='both')
ax.set_xticks(np.hstack((positions,positions+.25))+.1)
ax.set_xticklabels(xtick,fontsize=30)
ax.tick_params(axis='both', which='major', labelsize=30)
plt.title(f"Release: {A} vs. {B} ",fontsize=40)
plt.xlim(.75,1+len(names)*.6)
plt.ylabel('CFs',fontsize=40)
ax.yaxis.get_offset_text().set_size(30)
save_fig('release compare')
```

```python
##plotting box plots for Storage

positions =np.arange(1,1+len(names)*.6,.6)
fig=plt.figure(figsize=(40,20),dpi=300)
ax=fig.add_subplot(111,label='1')
plt.boxplot(A_list[1].T ,positions=positions,patch_artist=True,
           widths=.25,boxprops=dict(facecolor='b', color='b',alpha=.5,linewidth=5),
           medianprops=dict(color='k',linewidth=5),
       )
plt.boxplot(B_list[1].T ,positions=positions+.3,patch_artist=True,
           widths=.25,boxprops=dict(facecolor='r', color='r',alpha=.5,linewidth=5),
           medianprops=dict(color='g',linewidth=5)
       )
legend_smooth=mpatches.Patch(color='b',label=f'{A}')
legend_normal=mpatches.Patch(color='r',label=f'{B}')
ax.legend(handles=[legend_smooth, legend_normal], fontsize=40)
plt.grid(True,which='major',axis='both')
ax.set_xticks(np.hstack((positions,positions+.25))+.1)
ax.set_xticklabels(xtick,fontsize=30)
ax.tick_params(axis='both', which='major', labelsize=30)
plt.title(f"Storage: {A} vs. {B}",fontsize=40)
plt.xlim(.75,1+len(names)*.6)
plt.ylabel('AF',fontsize=40)
ax.yaxis.get_offset_text().set_size(30)
save_fig('Storage compare')
```
