```python
import pandas as pd
import numpy as np
import os
```

```python
##getting current dir

current = os.getcwd()
current
```

```python
##importing user data

user_data=pd.read_csv('input/user_data.csv')
n_stages=user_data.iloc[0,1]
user_data
```

```python
cascade_df=pd.read_excel('input/Cascade_info.xlsx',sheet_name='Cascade')
dam_name=cascade_df['name']
dam_name=sorted(dam_name)
dam_name
```

```python
##definnig input path and file names

path='output/node_solutions/main_tree/'
names=['total_pflow','storage','spill','release']
```

```python
##loading main tree

main_tree_path=os.path.join(current,'input','Tree.xlsx')
main_tree=pd.read_excel(main_tree_path)
```

```python
col_name=[]
for dam in dam_name:
    temp_name=dam+'_in'
    col_name.append(temp_name)
col_name=sorted(col_name)
col_name
```

```python
names
```

```python
col_name
```

```python
##making the solution trees

df_list=[]
co=['Node', 'Parent','Prob' ,'Stage','probability']
for n in names:
    file_p=os.path.join(current,path,"Nodes_"+n+'_solution_clean.csv')
    temp=pd.read_csv(file_p)
    if len(temp.columns)==len(dam_name)*2:
        temp.drop(columns=col_name,inplace=True)
    temp2=pd.concat([temp,main_tree[co]],axis=1)
    df_list.append(temp2)
```

```python
##checking on of them

df_list[1]
```

```python
##exporting solution trees

o_path="output/solution_tree/"
h=0
for tree in df_list:
    name_t="solution_tree_"+names[h]
    save_t=os.path.join(current,o_path,name_t+'.csv')
    tree.to_csv(save_t)
    h+=1
    print(save_t)
```

```python
df_list[1]
```

```python
full_list=[]
for solution in df_list:
    solution_list=[]
    for i in range(len(dam_name)):
        leaf=solution.loc[temp2['Stage']==n_stages].iloc[:,i]
        tmp=pd.DataFrame(index=range(n_stages),columns=range(len(leaf)))
        tmp.iloc[n_stages-1,:]=leaf
        parents=temp2.iloc[leaf.index,-4]
        for node in range(n_stages-2,-1,-1):
            tmp.iloc[node,:]=solution.iloc[parents-1,i]
            parents=temp2.iloc[parents-1,-4]
        solution_list.append(tmp)
    full_list.append(solution_list)
```

```python
o_path="output/solution_scenarios/"
s_name=['total_pflow','storage','spill','release']
d_name=sorted(dam_name)
s=0
for solution in full_list:
    d=0
    for dam in solution:
        name=d_name[d]+"_solution_scenarios_"+s_name[s]+'.csv'
#         print(name)
        save_p=os.path.join(current,o_path,name)
        print(save_p)
        dam.to_csv(save_p,index=False,header=False)
        d+=1
    s+=1
```
