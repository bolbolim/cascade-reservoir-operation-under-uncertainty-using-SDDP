```python
import numpy as np
import pandas as pd
import os
```

```python
##getting current path
folder_path="output/temp_files/node_solutions/Main_tree"
```

```python
cascade_df=pd.read_excel('input/Cascade_info.xlsx',sheet_name='Cascade')
name=cascade_df['name']
name
```

```python
##importing csv files

release_solution=pd.read_csv(os.path.join(folder_path,'Nodes_release_solution.csv'),header=None)
spill_solution=pd.read_csv(os.path.join(folder_path,'Nodes_spill_solution.csv'),header=None)
storage_solution=pd.read_csv(os.path.join(folder_path,'Nodes_storage_solution.csv'),header=None)
total_pflow_solution=pd.read_csv(os.path.join(folder_path,'Nodes_total_pflow_solution.csv'),header=None)

##making a list
solution_list_raw=[release_solution,spill_solution,storage_solution,total_pflow_solution]
```

```python
storage_solution
```

```python
##removing junk string

strings=['SDDP.State{Float64}\[','SDDP.State{Float64}\(','SDDP.State{Float64}\[SDDP.State{Float64}\(','\)','\)\]','\]','\[']
datasets=[]
for i in solution_list_raw:
    i=i.replace(strings,'',regex=True)
    i = i.astype(float)
    datasets.append(i)
# release_solution,spill_solution,storage_solution,total_pflow_solution=datasets
# solution_list=[release_solution,spill_solution,storage_solution,total_pflow_solution]
```

```python
col_name=[]
all_name=[]
for dam in name:
    temp_name=dam+'_out'
    temp_del=dam+'_in'
    all_name.append(temp_del)
    all_name.append(temp_name)
    col_name.append(temp_name)
```

```python
col_name
```

```python
##changing column names

for df in datasets:
    if df.shape[1]==len(name):
        df.columns=col_name
    elif df.shape[1]==len(name)*2:
        df.columns=all_name
    
```

```python
for df,f in zip(datasets,['release','spill','storage',"total_pflow"]):
    df=df.reindex(sorted(df.columns),axis=1)
    path=f'output/node_solutions/main_tree/Nodes_{f}_solution_clean.csv'
    df.to_csv(path,index=False)
```
